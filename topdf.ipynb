```python
import sys
import subprocess
import shutil
from pathlib import Path

try:
    import pdfkit
except ModuleNotFoundError:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'pdfkit'])
    import pdfkit

html_file = Path('technical_appendix.html')
pdf_file = html_file.with_suffix('.pdf')

if not html_file.exists():
    raise FileNotFoundError(f"Input HTML file not found: {html_file.resolve()}")

wkhtmltopdf = shutil.which('wkhtmltopdf')
if wkhtmltopdf is None:
    candidates = [
        Path(r'C:\Program Files\wkhtmltopdf\bin\wkhtmltopdf.exe'),
        Path(r'C:\Program Files (x86)\wkhtmltopdf\bin\wkhtmltopdf.exe'),
    ]
    wkhtmltopdf = next((str(path) for path in candidates if path.exists()), None)

if wkhtmltopdf is None:
    raise EnvironmentError(
        'wkhtmltopdf executable not found. Install wkhtmltopdf from https://wkhtmltopdf.org/downloads.html '
        'and make sure it is on your PATH or update the path in this notebook.'
    )

config = pdfkit.configuration(wkhtmltopdf=wkhtmltopdf)
pdfkit.from_file(str(html_file), str(pdf_file), configuration=config)
print(f'Converted {html_file.name} to {pdf_file.name}')
```
